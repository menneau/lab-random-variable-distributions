2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
num = pd.read_csv('numerical.csv')
```

```python
num
```

```python
num.columns
```

```python
cols_disc = ['months_since_last_claim', 'months_since_policy_inception', 'number_of_open_complaints', 'number_of_policies']
cols_cont = ['customer_lifetime_value', 'income', 'monthly_premium_auto', 'total_claim_amount']
```

```python
num_disc = num[cols_disc]
num_cont = num[cols_cont]
```

```python
num_disc_log = num_disc.apply(lambda x: np.where(x > 0, np.log(x), x))
num_cont_log = num_cont.apply(lambda x: np.where(x > 0, np.log(x), x))
```

# Continuous (except income)

```python
sns.displot(num_cont['customer_lifetime_value'], kde=True)
sns.displot(num_cont['monthly_premium_auto'], kde=True)
sns.displot(num_cont['total_claim_amount'], kde=True)
```

Continuous(except income) log transformed.

```python
sns.displot(num_cont_log['customer_lifetime_value'], kde=True)
sns.displot(num_cont_log['monthly_premium_auto'], kde=True)
sns.displot(num_cont_log['total_claim_amount'], kde=True)
```

### Conclusions.

- Customer lifetime value, monthly premium auto and total claim amount have different peaks. Most of the values don't seem to be concentrated around the mean, so they don't seem to fit normal or exponential distributions.
- Applying the log transformation helps reducing the right-hand tails.
- Total claim amount could have some similarity with an exponential distribution after the log transformation.

## Continuous. Income.

```python
sns.displot(num_cont['income'], kde=True)
```

```python
sns.displot(num_cont_log['income'], kde=True)
```

We remove the zero values (data from unemployed customers).

```python
income = num_cont['income']
income_pos = income[income>0]
```

```python
sns.displot(income_pos, kde=True)
```

```python
income_pos_log = income_pos.apply(lambda x: np.where(x > 0, np.log(x), x))
```

```python
sns.displot(income_pos_log, kde=True)
```

### Conclusions.

- We can see the same kind of problem with income.
- It has different peaks. Most of the values don't seem to be concentrated around the mean, so it doesn't seem to fit a normal or an exponential distribution.

### Conclusions.
